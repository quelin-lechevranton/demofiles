# Central limit theorem

Let $\{Y_i\}_{1\leq i \leq N}$ a set of random variables, independant and identically distributed (iid) with mean $\mu$ and variance $\sigma^2$, then the probability distribution function of the sample mean:
$$
\bar{X} = \frac{1}{N} \sum_i Y_i
$$
converges as $N \rightarrow \infty$ to a gaussian distribution of mean $\mu$ and variance $\sigma^2/N$.


# Let's verify that, first with uniform random variables.

```python
%matplotlib inline
pl.rcParams['figure.figsize'] = (10.0, 8.0)
pl.rcParams['font.size'] = 18
```

```python
N_realizations = 10000
N = 1

y = np.zeros(N_realizations)
for r in range(N_realizations):
    x = 2 * np.random.rand(N) - 1
    y[r] = np.mean(x)

_ = pl.hist(y, bins=100, density=1)

sig = 1 / np.sqrt(3.*N)
xx = np.linspace(-5.*sig, 5. * sig, 100)
yy = 1 / (np.sqrt(2. * np.pi) * sig) * np.exp(-0.5 * xx**2 / sig**2)
pl.plot(xx, yy, 'r--')
```

# We can start with more esoteric random variables

```python
import scipy.stats
```

```python
bins = np.linspace(-1, 1., 21)
a = np.zeros(len(bins)-1)
a[:3] = 1.
a[-3:] = 1.
```

```python
dist = scipy.stats.rv_histogram((a,bins))
```

```python
X = dist.rvs(size=1000000)
_ = pl.hist(X, bins=100)
```

```python
N=80
y = np.zeros(N_realizations)
for r in range(N_realizations):
    x = dist.rvs(size=N)
    y[r] = np.mean(x)

_ = pl.hist(y, bins=100, density=1)

sig = np.sqrt(0.73/N)
xx = np.linspace(-5.*sig, 5. * sig, 100)
yy = 1 / (np.sqrt(2. * np.pi) * sig) * np.exp(-0.5 * xx**2 / sig**2)
pl.plot(xx, yy, 'r--')
```
